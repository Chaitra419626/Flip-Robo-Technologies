# Checking for the presence of outliers in our dataset

```python
plt.figure(figsize = (40, 30))
plotnumber = 1

for col in df1.columns:
    if plotnumber <= 70:
        ax = plt.subplot(14, 5, plotnumber)
        sns.boxplot(df1[col])
        plt.xlabel(col, fontsize = 15)
    
    plotnumber += 1
plt.tight_layout()
plt.show()
```

**We see the presence of outliers in our dataset, now let's check for skewness**

```python
df1.skew()
```
